# PCA Formulae Revision
A small real-life example with only formulas and manual calculation.

We take 3 students and 3 features:
- Student A = (study hours = 2, test score = 40, assignment score = 45)
- Student B = (study hours = 4, test score = 60, assignment score = 65)
- Student C = (study hours = 6, test score = 80, assignment score = 85)

Here the 3 features are related.
A student who studies more is also getting better scores.
PCA will combine this repeated information into new principal components.

For short manual calculation below, we use only the last 2 features:
- A = (40, 45)
- B = (60, 65)
- C = (80, 85)

This keeps the math short, but the idea is the same for 3 features also.

## Step 1: Mean of Each Feature
Why this step?
We find the center of the data first.
PCA measures spread around the center, not around raw values.

For feature 1:
$$
\mu_1 = \frac{40 + 60 + 80}{3} = 60
$$

For feature 2:
$$
\mu_2 = \frac{45 + 65 + 85}{3} = 65
$$

Mean vector:
$$
\mu = (60, 65)
$$

## Step 2: Center the Data
Why this step?
Centering shifts data around mean 0.
This makes PCA focus on variation, not absolute magnitude.

Formula:
$$
X_{centered} = X - \mu
$$

Centered points:
- A' = (40-60, 45-65) = (-20, -20)
- B' = (60-60, 65-65) = (0, 0)
- C' = (80-60, 85-65) = (20, 20)

Centered matrix:
$$
X = \begin{bmatrix}
-20 & -20 \\
0 & 0 \\
20 & 20
\end{bmatrix}
$$

## Step 3: Covariance Matrix
Why this step?
Covariance tells how features change together.
PCA uses this matrix to find the main spread directions.

Formula:
$$
\text{Cov}(X) = \frac{1}{n} X^T X
$$

Here,
$$
X^T X = \begin{bmatrix}
800 & 800 \\
800 & 800
\end{bmatrix}
$$

So,
$$
\text{Cov}(X) = \frac{1}{3}\begin{bmatrix}
800 & 800 \\
800 & 800
\end{bmatrix}
= \begin{bmatrix}
\frac{800}{3} & \frac{800}{3} \\
\frac{800}{3} & \frac{800}{3}
\end{bmatrix}
$$

## Step 4: Eigenvalues and Eigenvectors
Why this step?
Eigenvalues tell how much variance each direction captures.
Eigenvectors give the directions of the principal components.

For the covariance matrix
$$
A = \begin{bmatrix}
\frac{800}{3} & \frac{800}{3} \\
\frac{800}{3} & \frac{800}{3}
\end{bmatrix}
$$

We solve:
$$
\det(A - \lambda I) = 0
$$

So,
$$
\det\begin{bmatrix}
\frac{800}{3} - \lambda & \frac{800}{3} \\
\frac{800}{3} & \frac{800}{3} - \lambda
\end{bmatrix} = 0
$$

$$
\left(\frac{800}{3} - \lambda\right)^2 - \left(\frac{800}{3}\right)^2 = 0
$$

$$
(-\lambda)\left(\frac{1600}{3} - \lambda\right) = 0
$$

Eigenvalues:
$$
\lambda_1 = \frac{1600}{3}, \qquad \lambda_2 = 0
$$

Eigenvectors:
$$
v_1 = \frac{1}{\sqrt{2}}\begin{bmatrix}1 \\ 1\end{bmatrix}, \qquad
v_2 = \frac{1}{\sqrt{2}}\begin{bmatrix}1 \\ -1\end{bmatrix}
$$

So:
- PC1 is along $(1,1)$
- PC2 is along $(1,-1)$

## Step 5: Final Principal Components
Why this step?
This step converts original features into new PCA features.
We keep the components that capture most variance.

Principal component formulas:
$$
PC1 = \frac{1}{\sqrt{2}}X_1 + \frac{1}{\sqrt{2}}X_2
$$

$$
PC2 = \frac{1}{\sqrt{2}}X_1 - \frac{1}{\sqrt{2}}X_2
$$

Scores for centered points:
- For (-20, -20): PC1 = $-20\sqrt{2}$, PC2 = 0
- For (0, 0): PC1 = 0, PC2 = 0
- For (20, 20): PC1 = $20\sqrt{2}$, PC2 = 0

Explained variance ratio:
$$
\frac{\lambda_1}{\lambda_1 + \lambda_2} = \frac{1600/3}{1600/3 + 0} = 1
$$

So PC1 captures 100% variance and PC2 captures 0%.

## Revision Summary
- Mean centers the data
- Centering removes location effect
- Covariance shows joint variation
- Eigenvector with largest eigenvalue gives PC1
- Keep top PCs that capture most variance

## 3-Feature Real-Life Intuition
If we used all 3 features directly:
- study hours
- test score
- assignment score

Then PCA would create:
- PC1 = overall academic performance
- PC2 = a smaller leftover pattern
- PC3 = another even smaller leftover pattern

So PCA does not pick old columns.
It creates new combined columns from the old features.
