# Simple Outlier Example
This notebook creates a small dataset with a few extreme values and plots it so the outliers are easy to see.

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame(
    {
        "value": [10, 11, 12, 11, 10, 13, 12, 11, 10, 65, 72]
    }
)

print(data)

plt.figure(figsize=(8, 4))
plt.plot(data.index, data["value"], marker="o", linestyle="-")
plt.title("Simple Data with Outliers")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True, alpha=0.3)
plt.show()
```
