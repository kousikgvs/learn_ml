# Feature Scaling - Normalization

Normalization is a feature scaling technique used to bring numerical values into a common range. It helps different features stay comparable without changing the meaning of the data.

Normalization is usually used when values need to be scaled between 0 and 1. It is useful for models that depend on distance, such as KNN and K-Means.

## Formula

$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$

Where:
- $x$ = original value
- $x_{min}$ = minimum value in the feature
- $x_{max}$ = maximum value in the feature
- $x_{norm}$ = normalized value
