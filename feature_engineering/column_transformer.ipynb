# Column Transformer

`ColumnTransformer` lets us apply different preprocessing steps to different columns in the same dataset.

In this example, we will:
- scale numerical columns
- one-hot encode categorical columns
- combine everything into one transformed output

This is useful when a dataset contains both numbers and text categories.

```python
# Import required libraries
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
# Create a sample dataset with numerical and categorical columns
data = pd.DataFrame({
    'age': [22, 25, 47, 52, 46],
    'salary': [25000, 32000, 52000, 61000, 58000],
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Mumbai'],
    'education': ['UG', 'PG', 'PG', 'UG', 'PhD']
})

data
```

## Apply Different Transformations

We will use:

- `StandardScaler()` for `age` and `salary`
- `OneHotEncoder()` for `city` and `education`

`ColumnTransformer` combines these transformations into one preprocessing step.

```python
# Define which columns belong to each group
numerical_columns = ['age', 'salary']
categorical_columns = ['city', 'education']

# Create the column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_columns),
        ('cat', OneHotEncoder(sparse_output=False), categorical_columns)
    ]
)

# Fit and transform the dataset
transformed_data = preprocessor.fit_transform(data)

# Get output feature names
feature_names = preprocessor.get_feature_names_out()

transformed_df = pd.DataFrame(transformed_data, columns=feature_names)
transformed_df
```

```python
# Show the transformed shape and feature names
print('Original shape:', data.shape)
print('Transformed shape:', transformed_df.shape)
print('Feature names:')
for feature_name in feature_names:
    print(feature_name)
```
