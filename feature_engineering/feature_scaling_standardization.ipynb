# Feature Scaling - Standardization

Feature scaling is the process of bringing numerical features to a similar range or scale. It helps machine learning models treat all features more fairly during training.

### Standardization

Standardization transforms data so that it has a mean of 0 and a standard deviation of 1. It is also called z-score normalization and is useful when the data has outliers or follows a roughly normal distribution.

### Formula

$z = \frac{x - \mu}{\sigma}$

Where:
- $x$ = original value
- $\mu$ = mean of the feature
- $\sigma$ = standard deviation of the feature
- $z$ = standardized value

## Standardization

Standardization changes feature values so that the data is centered around 0 with unit variance. This makes features comparable without forcing them into a fixed minimum and maximum range.


## Standardization and Outliers

Standardization does not remove outliers. It only changes the scale of the data by using the mean and standard deviation.

If a feature has very large outliers, those outliers can still affect the mean and standard deviation. So standardization is influenced by outliers, even though it puts the values on a common scale.

### When Standardization Helps

Standardization is useful for models that depend on distance, magnitude, or gradient updates. It often improves the performance of K-Means, KNN, PCA, ANN, Logistic Regression, and Gradient Descent based models.

### Models That Usually Do Not Need It

Tree-based models such as Decision Tree, Random Forest, Gradient Boosting, and XGBoost usually do not need standardization. These models split data based on feature thresholds, so scale is usually not important.
