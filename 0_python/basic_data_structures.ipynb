## 1) Data Types
Python has multiple built-in data types: numbers, strings, booleans, lists, tuples, sets, dictionaries, and `NoneType`.

</div>

```python
def demo_data_types() -> None:
    samples = {
        "int": 42,
        "float": 3.14,
        "complex": 2 + 3j,
        "str": "hello",
        "bool": True,
        "list": [1, 2, 3],
        "tuple": (1, 2, 3),
        "set": {1, 2, 3},
        "dict": {"name": "Alice", "age": 25},
        "none": None,
    }

    for name, value in samples.items():
        print(f"{name:>7}: {value!r:<20} -> {type(value).__name__}")


demo_data_types()
```

<div style="font-size:16px; line-height:1.6;">

## 2) Operators
Python supports arithmetic, comparison, logical, assignment, and membership operators.

</div>

```python
def demo_operators(a: int, b: int) -> None:
    print("Arithmetic:")
    print(f"a + b = {a + b}")
    print(f"a - b = {a - b}")
    print(f"a * b = {a * b}")
    print(f"a / b = {a / b}")
    print(f"a // b = {a // b}")
    print(f"a % b = {a % b}")
    print(f"a ** b = {a ** b}")

    print("\nComparison:")
    print(f"a > b: {a > b}")
    print(f"a == b: {a == b}")

    print("\nLogical:")
    print((a > 0) and (b > 0))

    print("\nMembership:")
    print(f"2 in [1, 2, 3]: {2 in [1, 2, 3]}")


demo_operators(10, 3)
```

<div style="font-size:16px; line-height:1.6;">

## 3) Tuples and Lists
- **List**: mutable, ordered
- **Tuple**: immutable, ordered

</div>

```python
def demo_tuples_lists() -> None:
    fruits = ["apple", "banana", "cherry"]
    coords = (10, 20)

    fruits.append("mango")
    print("List:", fruits)
    print("Tuple:", coords)
    print("First fruit:", fruits[0])
    print("Tuple unpacking:", coords[0], coords[1])


demo_tuples_lists()
```

<div style="font-size:16px; line-height:1.6;">

## 4) Set Operators
Sets are unordered collections of unique elements.
Common operators: union (`|`), intersection (`&`), difference (`-`), symmetric difference (`^`).

</div>

```python
def demo_set_operators() -> None:
    a = {1, 2, 3, 4}
    b = {3, 4, 5, 6}

    print("Union:", a | b)
    print("Intersection:", a & b)
    print("Difference (a-b):", a - b)
    print("Symmetric Difference:", a ^ b)


demo_set_operators()
```

<div style="font-size:16px; line-height:1.6;">

## 5) Dictionaries
Dictionaries store key-value pairs and are mutable.

</div>

```python
def demo_dictionaries() -> None:
    student = {"name": "Kiran", "age": 22, "course": "ML"}
    student["age"] = 23
    student["city"] = "Hyderabad"

    print("Dictionary:", student)
    print("Keys:", list(student.keys()))
    print("Values:", list(student.values()))
    print("Items:", list(student.items()))


demo_dictionaries()
```

## 6) Conditionals
Use `if`, `elif`, and `else` for decision making.

</div>

```python
def classify_number(num: int) -> str:
    if num > 0:
        return "Positive"
    if num < 0:
        return "Negative"
    return "Zero"


for value in [10, -3, 0]:
    print(value, "->", classify_number(value))
```

<div style="font-size:16px; line-height:1.6;">

## 7) Iteratives (Loops)
Use `for` and `while` loops to repeat tasks.

</div>

```python
def demo_iteratives() -> None:
    print("For loop:")
    for i in range(1, 4):
        print("Iteration", i)

    print("\nWhile loop:")
    n = 3
    while n > 0:
        print("Countdown:", n)
        n -= 1


demo_iteratives()
```

<div style="font-size:16px; line-height:1.6;">

## 8) Yield Functions / Generator Functions
A generator uses `yield` to produce values one at a time, saving memory.

</div>

```python
def even_numbers(limit: int):
    for i in range(limit + 1):
        if i % 2 == 0:
            yield i


print("Generated evens:", list(even_numbers(10)))
```

<div style="font-size:16px; line-height:1.6;">

## 9) List Comprehension
List comprehensions provide a concise way to build lists.

</div>

```python
def demo_list_comprehension() -> None:
    squares = [x * x for x in range(1, 6)]
    even_squares = [x * x for x in range(1, 11) if x % 2 == 0]
    print("Squares:", squares)
    print("Even squares:", even_squares)


demo_list_comprehension()
```

## 10) Important Built-in Functions
This section demonstrates:
`len()`, `range()`, `zip()`, `enumerate()`, `map()`, `filter()`, `sorted()`

</div>

```python
def square(x: int) -> int:
    return x * x


def is_even(x: int) -> bool:
    return x % 2 == 0


def demo_builtin_functions() -> None:
    nums = [5, 2, 9, 1, 6]
    names = ["Alice", "Bob", "Charlie"]
    scores = [85, 90, 95]

    print("len(nums):", len(nums))
    print("range(5):", list(range(5)))
    print("zip(names, scores):", list(zip(names, scores)))
    print("enumerate(names):", list(enumerate(names, start=1)))
    print("map(square, nums):", list(map(square, nums)))
    print("filter(is_even, nums):", list(filter(is_even, nums)))
    print("sorted(nums):", sorted(nums))
    print("sorted(nums, reverse=True):", sorted(nums, reverse=True))


demo_builtin_functions()
```
