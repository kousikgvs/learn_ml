# Real-Life Applications of Probability



Probability is used whenever we need to measure uncertainty, estimate risk, or make decisions with incomplete information.



## Common Real-Life Applications



1. **Weather Forecasting**  

Meteorologists use probability to estimate the chance of rain, storms, or extreme weather.  

Example: A 70% chance of rain means rain is likely, but not guaranteed.



2. **Medical Diagnosis and Testing**  

Doctors use probability to interpret test results, disease risk, and treatment outcomes.  

Example: The probability that a patient actually has a disease after a positive test depends on both test accuracy and disease prevalence.



3. **Finance and Insurance**  

Banks, investors, and insurance companies use probability to estimate losses, returns, and default risk.  

Example: An insurance company calculates the probability of accidents to decide premium amounts.



4. **Quality Control in Manufacturing**  

Factories use probability to estimate defect rates and maintain product quality.  

Example: If 2 out of every 100 items are usually defective, probability helps predict how many faulty items may appear in a batch.



5. **Sports Analytics**  

Teams and analysts use probability to predict outcomes and improve strategies.  

Example: The probability of a team winning can be estimated using past performance, player statistics, and match conditions.



6. **Games and Gambling**  

Probability explains outcomes in dice, cards, lotteries, and casino games.  

Example: The probability of drawing an ace from a standard deck is 4/52 = 1/13.



7. **Machine Learning and Artificial Intelligence**  

Probability is used in prediction models, classification, and uncertainty estimation.  

Example: A spam filter may assign a probability that an email is spam before deciding whether to block it.



8. **Traffic and Transportation**  

Probability helps estimate travel time, traffic congestion, and accident likelihood.  

Example: Navigation apps use past and real-time data to estimate the probability of delays on a route.



9. **Elections and Polling**  

Pollsters use probability to estimate how likely a candidate is to win.  

Example: Survey samples are used to estimate voting trends in the full population.



10. **Supply Chain and Business Forecasting**  

Businesses use probability to predict demand, inventory needs, and delivery delays.  

Example: A store estimates the probability that demand for a product will rise during a festival season.



## Short Conclusion



In real life, probability helps people make better decisions under uncertainty. It is widely used in science, business, health, technology, and everyday planning.


## Modular Probability Code in Python

The following code organizes common probability operations into small reusable functions. This makes the notebook easier to read, test, and extend.

```python
from math import factorial


def validate_probability(value: float) -> float:
    if not 0 <= value <= 1:
        raise ValueError("Probability must be between 0 and 1.")
    return value



def factorial_non_negative(n: int) -> int:
    if n < 0:
        raise ValueError("n must be a non-negative integer.")
    return factorial(n)



def permutations(n: int, r: int) -> int:
    if not 0 <= r <= n:
        raise ValueError("r must satisfy 0 <= r <= n.")
    return factorial(n) // factorial(n - r)



def combinations(n: int, r: int) -> int:
    if not 0 <= r <= n:
        raise ValueError("r must satisfy 0 <= r <= n.")
    return factorial(n) // (factorial(r) * factorial(n - r))



def probability(favorable_outcomes: int, total_outcomes: int) -> float:
    if total_outcomes <= 0:
        raise ValueError("total_outcomes must be greater than 0.")
    if favorable_outcomes < 0 or favorable_outcomes > total_outcomes:
        raise ValueError("favorable_outcomes must be between 0 and total_outcomes.")
    return favorable_outcomes / total_outcomes



def conditional_probability(intersection_probability: float, given_probability: float) -> float:
    validate_probability(intersection_probability)
    validate_probability(given_probability)
    if given_probability == 0:
        raise ValueError("given_probability must not be 0.")
    return intersection_probability / given_probability



def complement_probability(event_probability: float) -> float:
    validate_probability(event_probability)
    return 1 - event_probability



def independent_joint_probability(*probabilities: float) -> float:
    joint = 1.0
    for value in probabilities:
        joint *= validate_probability(value)
    return joint



def bayes_theorem(prob_b_given_a: float, prob_a: float, prob_b: float) -> float:
    validate_probability(prob_b_given_a)
    validate_probability(prob_a)
    validate_probability(prob_b)
    if prob_b == 0:
        raise ValueError("prob_b must not be 0.")
    return (prob_b_given_a * prob_a) / prob_b
```

```python
examples = {
    "P(rolling an even number on a die)": probability(3, 6),
    "5P2": permutations(5, 2),
    "5C2": combinations(5, 2),
    "Complement of 0.65": complement_probability(0.65),
    "Joint probability of two independent events (0.5 and 0.3)": independent_joint_probability(0.5, 0.3),
    "Conditional probability P(A|B) with P(A∩B)=0.12 and P(B)=0.4": conditional_probability(0.12, 0.4),
    "Bayes theorem result": bayes_theorem(0.9, 0.01, 0.05),
}

examples
```
